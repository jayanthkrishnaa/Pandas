#  Handling Missing Values

```python
import pandas as pd
```

```python
df = pd.read_csv("weather.csv",parse_dates = ["day"])
df
```

## set_index()

```python
df.set_index('day',inplace = True) #Here the day column is set as an index for the dataframe
df
```

# fillna() 

```python
df.fillna(0) # replaces all the NaN values with the value provided as the function argument,here it is 0.
```

```python
n_df = df.fillna({'temparature': 0,
                 'windspeed': 0,        #Here we are passing the dictionary where the values in it are replaced with NaN
                 'event': 'NO EVENT'})
n_df
```

```python
n_df = df.fillna(method = 'ffill')  #ffill = foward fill; the same values are filled in the following empty cell.
n_df
```

```python
n_df = df.fillna(method = 'bfill')
n_df
```

```python
n_df = df.fillna(method = 'ffill',axis = 'columns')
n_df
```

```python
n_df = df.fillna(method = 'ffill',limit =1) #It limits the forward filling to only one cell
n_df
```

## Interpolate

```python
n_df = df.interpolate() #It fills the NaN cells with some related value based on the dataset
n_df
```

```python
n_df = df.interpolate(method = 'time')
#method has various values that it can take, here we have used 'time' for our interpoaltion
n_df
```

# dropna()

```python
n_df = df.dropna() #drops all the rows that contian the NaN values.
n_df
```

```python
n_df = df.dropna(how = 'all') # it drops the row that contain the NaN values in all the columns.
n_df
```

```python
n_df  = df.dropna(thresh = 1) #it keeps the threshold of 1 for the NaN in the cells.
n_df
```

# Inserting the missing values 

```python
dt = pd.date_range('01-01-2017','01-11-2017')
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)             #reindexing the dataframe by including the missing date in it.
df
```

# Replace 

```python
import numpy as np
```

```python
df = pd.read_csv('weather_data.csv')
df
```

```python
n_df = df.replace(-99999,np.NaN)  #Here the value -99999 is replaced with the NaN
n_df
```

```python
n_df = df.replace([-99999,2],np.NaN) #Here the list of values is used to replace them.
n_df
```

```python
n_df = df.replace({'temperature': -99999,
                  'windspeed': -99999,      #Here the dictionay of vlaues are passed that need to be replaced.
                  'event': '0'},np.NaN)
n_df
```

```python
n_df = df.replace({-99999: np.NaN, # Here the symbol ':' is used to map the values in the dataframe cells.
                   '0':'Sunny'})
n_df
```

# Using Regex 

```python
df = pd.read_csv("weather_data.csv")
df
```

```python
n_df = df.replace('[A-Za-z]','',regex = True)
n_df
```

```python
n_df = df.replace({'temperature': '[A-Za-z]',
                  'windspeed': '[A-Za-z]'},'',regex = True)
n_df
```

#### One list of values replacing the other
